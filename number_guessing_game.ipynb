```python
import random
secret_number = random.randint(1,100)
for i in range(0,9):
    user_input=int(input("Enter the number :"))
    try:
        if user_input<1 or user_input>100:
            print("invalid number..!")
        elif user_input < secret_number:
            print(f"{user_input} is less than the secret_number")
        elif user_input > secret_number:
            print(f"{user_input} is greater than the secret_number")
        else:
            print("you have guessed the correct number")
            break
    except ValueError:
            print("Enter only the numbers from 1-100")
            continue
if  user_input== secret_number:
    print(f"You have guessed the secret number in {i} attemts..")
else:
    print("try again..!")
```
